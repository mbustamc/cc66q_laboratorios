  Bag of Words analiza la frecuencia absoluta de las palabras en un documento, y a partir de ella se busca entender el significado o contexto de éste

```python
# Leer el archivo
with open('data/libros.txt', 'r', encoding='utf-8') as archivo:
    libros = archivo.readlines()
libros = [libro.strip() for libro in libros]  # Se eliminan saltos de línea

# Transformación de strings (mayúsculas)
libros = [libro.upper().split(' - ') for libro in libros]

# Búsqueda de strings:
palabras_clave = 'LA'
titulos_tokenizados = [libro[0].split() for libro in libros] # Se tokenizan los títulos

libros_con_palabra_clave = []
for titulo in titulos_tokenizados:
    for token in titulo:
        if token == palabras_clave:
            libros_con_palabra_clave.append(' '.join(titulo)) # Se regenera el título original
            break # Termina el loop interno si se encuentra la palabra clave

print("libros_con_palabra_clave:", libros_con_palabra_clave)

# Split de strings:
primer_nombre_autores = [libro[1].split()[0] for libro in libros]
print("primer_nombre_autores:", primer_nombre_autores)

# Longitud de strings:
longitud_titulos = [len(titulo_tokens) for titulo_tokens in titulos_tokenizados]
for i in range(len(longitud_titulos)):
    print(f"'{libros[i][0]}': {longitud_titulos[i]} palabras.")
```

```python
import nltk
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt


from collections import Counter



# Se descargan los corpus necesarios
nltk.download('gutenberg')
nltk.download('stopwords')

from nltk.corpus import gutenberg
from nltk.corpus import stopwords

# Carga el texto "Alicia en el país de las maravillas"
texto = gutenberg.raw('carroll-alice.txt')
# Stopwords en inglés
stopwords_en = set(stopwords.words('english'))


#Definimos una función para normalizar texto
def normalize(text, remove_tildes = True):
    """Normaliza una cadena de texto convirtiéndo todo a minúsculas, quitando los caracteres no alfabéticos y los tildes"""
    text = text.lower() # Llevamos todo a minúscula
    text = re.sub(r'[^A-Za-zñáéíóú]', ' ', text) # Reemplazamos los caracteres no alfabéticos por un espacio
    if remove_tildes:
        text = re.sub('á', 'a', text) # Reemplazamos los tildes
        text = re.sub('é', 'e', text)
        text = re.sub('í', 'i', text)
        text = re.sub('ó', 'o', text)
        text = re.sub('ú', 'u', text)
    return text

texto_normalizado = normalize(texto, remove_tildes=False)
texto_normalizado_tokenizado = texto_normalizado.split()


# Crear el wordcloud
nube = WordCloud(width=800, height=400, background_color='white',
                 stopwords=stopwords_en).generate(texto_normalizado)

# Mostrarlo
plt.imshow(nube.generate_from_frequencies(
    {word: f for word, f in Counter(texto_normalizado_tokenizado).items() if not word in stopwords_en}), interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### 1. Tokenizador

Como vimos en clases pasadas, el como separar los principales token de una oración no siempre es una tarea fácil. Por lo que para separar los principales token del texto asociado a cada sentimiento se pide que defina un tokenizador, que entregue una lista de los principales tokens. Su tokenizador debe contener al menos tres expresiones regulares, que usted estime conveniente para hacer la separación de los token dentro de la oración, para ello complete le siguiente función:

```python
import re

def tokenizador(oracion):

    patrones = [
        r'https?://[^\s]+',  # URLs
        r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]+',  # emails
        r'#[A-Za-z0-9_]+',  # hashtags
        r'@[A-Za-z0-9_]+',  # menciones
        r"[A-Za-záéíóúüñÁÉÍÓÚÜÑ]+(?:'[A-Za-záéíóúüñÁÉÍÓÚÜÑ]+)?", #palabras con tildes
        r'\d+(?:[.,]\d+)?'  # numeros
    ]
    
    # Combinar todos los patrones
    patron_combinado = '|'.join(patrones)
    
    # Encontrar todos los tokens
    tokens = re.findall(patron_combinado, oracion)
    
    # Limpieza
    tokens = [t.strip() for t in tokens if t.strip()]
    return tokens
```
